# Simple plotter for fispact results

```python
 %matplotlib inline
# imports
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import fispact_analysis as fa
import fispact_output_reader
```

```python
# data to selct the right file
material = "Al"
p_energy = 28.0  # energy in MeV
irrad_time = 300 # irradiation time in seconds  3600, 900 and 300 are simulated in fispact
```

```python
# get the fispact output 
ef = str(p_energy) + "MeV/"
fn1 = material + "_" + str(p_energy)
if irrad_time == 3600:
    tme_str = "1hr/"
elif irrad_time == 900:
    tme_str = "15min/"
else:
    tme_str = "5min/"
path = "simulations/fispact/"+ tme_str + material + "/TENDL/" + fn1 + ".out"
fout = fispact_output_reader.read_fis_out(path)
```

```python
# plot activity over time
fa.plot_act(fout.sumdat)
```

```python
# plots for a single cooling time
ts_10min = fout.timestep_data[2]  # need to make it link to cooling time - current set for 10mins
# plot the fispact calculated 27 group gamma spec
fa.plot_spectra(ts_10min)
```
